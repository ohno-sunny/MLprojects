Linear regression - two training examples & one feature

We have two training examples as below.

Q. What would be the temperature after 6 hours of heating?

| Duration of heating(hr) | Temperature('C) |
|--------------------|------------|
| 3.0                | 30         |
| 4.0                | 50         |

```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
x_train = np.array([3.0, 4.0])
y_train = np.array([30.0, 50.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# counting the number of training examples
m = x_train.shape[0] #add [0] because without it, it's(2,)
print(f"Number of training examples is: {m}")
print(f"x_train.shape : {x_train.shape}")
print(f"Number of training exaples is: {len(x_train)}")
```

```python
# Plotting these points
plt.scatter(x_train, y_train, marker='x', c='r')
```

Linear regression

$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b $$

Cost function to see the deviation from my prediction

  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 $$

```python
# cost fuction
# checing to see if there is an error
def compute_cost (x, y, w, b):
    m = x.shape[0]
    cost_sum = 0
    for i in range (m):
      f_wb = w * x[i] + b
      cost = (f_wb - y[i])**2
      cost_sum = cost_sum + cost
    total_cost = (1 / (2 * m)) * cost_sum

    return total_cost
```

```python
a = compute_cost (x=x_train, y=y_train, w=20, b=-30)
print (a)
```

```python
x_i = 6
prediction = w * x_i + b
print (f"It will be {prediction} degrees celcius")
```
